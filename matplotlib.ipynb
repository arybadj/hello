```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,5,5]
y=[1,2,3,5,5]
plt.plot(x,y)

plt.legend("mje") # blue line ky reprent krti h yeh btata h
plt.title("my first plot")  # to give title
plt.xlabel("power") # giving nmae to the axis
plt.ylabel("weak")
plt.show()
```

```python
x=[1,2,3,5,5]
y=[1,2,3,5,5]
plt.plot(x,y, lw=7)  ## increase length width using lw command

plt.legend("mje") 
plt.title("my first plot")  
plt.xlabel("power") 
plt.ylabel("weak")
plt.show()
```

```python
x=[1,2,3,5,5]
y=[1,2,3,5,5]
plt.plot(x,y,lw=5, ls='dotted')  # ls is use to describe the type of line

plt.legend("mje") 
plt.title("my first plot")  
plt.xlabel("power") 
plt.ylabel("weak")
plt.show()
```

```python
x=[1,2,13,5,5]
y=[1,2,3,5,15]
plt.plot(x,y,lw=5, ls='dotted',marker='*')  # describe changes

plt.legend("mje") 
plt.title("my first plot")  
plt.xlabel("power") 
plt.ylabel("weak")
plt.show()
```

```python
x=[1,2,13,5,5]
y=[1,2,32,5,15]
plt.plot(x,y,lw=5, ls='dotted',marker='*',ms=14,mfc='r',mec='y')  # ms = size of marker mfc maker colour mec fills outer colour

plt.legend("mje") 
plt.title("my first plot")  
plt.xlabel("power") 
plt.ylabel("weak")
plt.show()
```
